```python
import pandas as pd

# Function to fetch and display stock details
def get_stock_details(user):
    file_path = r"Stock_Register.xlsx"  # Path to the Excel file

    # Load the Excel file into a DataFrame
    try:
        df = pd.read_excel(file_path)
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
        return
    except Exception as e:
        print(f"An error occurred while reading the Excel file: {e}")
        return

    # Column names
    try:
        product_id = 'Product ID'
        product_name = 'Product Name'
        quantity_available = 'Quantity Available'

        productID = df[product_id]
        productName = df[product_name]
        quantityAvailable = df[quantity_available]
    except KeyError as e:
        print(f"Error: Missing column in the Excel file - {e}")
        return

    # Logic for Owner
    if user == 'Owner':
        print("Product IDs:\n", productID)
        print("Product Names:\n", productName)
        print("Quantities Available:\n", quantityAvailable)
    else:  # Logic for other users
        print("Product Names and Availability:")
        for name, quantity in zip(productName, quantityAvailable):
            availability = "Available" if quantity > 0 else "Not Available"
            print(f"{name}: {availability}")

# Get user input and fetch stock details
user = input("Enter the user: ")
get_stock_details(user)
```

```python
import pandas as pd

def get_customers_in_time_range(excel_file, start_time, end_time):
    # Read data from Excel file
    df = pd.read_excel(excel_file)

    # Filter data based on time range
    filtered_df = df[(df['Time'] >= start_time) & (df['Time'] <= end_time)]

    # Calculate total number of customers
    total_customers = filtered_df['Number of Customers'].sum()

    return total_customers


# Example usage
excel_file =str(input('Enter the excel_file name '))
start_time = input("Enter start time (HH:MM): ")
end_time = input("Enter end time (HH:MM): ")

total_customers = get_customers_in_time_range(excel_file, start_time, end_time)

print(f"Total customers between {start_time} and {end_time}: {total_customers}")
```

```python
import pandas as pd

def get_consumptions_in_time_range(excel_file, start_time, end_time):
    # Read data from Excel file
    df = pd.read_excel(excel_file)

    # Filter data based on time range
    filtered_df = df[(df['Time'] >= start_time) & (df['Time'] <= end_time)]

    # Calculate total number of customers
    total_consumption = filtered_df['Power Consumption (Watts)'].sum()

    return total_consumption


# Example usage
excel_file =str(input('Enter the excel_file name'))
start_time = input("Enter start time (HH:MM): ")
end_time = input("Enter end time (HH:MM): ")

total_consumptions = get_consumptions_in_time_range(excel_file, start_time, end_time)

print(f"Total customers between {start_time} and {end_time}: {total_consumptions}")
```
