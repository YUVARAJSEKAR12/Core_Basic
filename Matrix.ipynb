```python
import numpy as np#sort
x = np.array([[10,2,3,4],[20,4,5,6],[10,20,39,3]])
print("Original mat:")
print(x)

y = np.min(x)
z=np.max(x)
a=np.mean(x)
print("Min",y)
print("Average",a)
print("max",z)
```

```python
import numpy as np#reverse
x = np.array([[1,2,3,4],[2,4,5,6],[10,20,39,3]])
print("Original mat:")
print(x)
print("Reverse mat:")
y = x[::-1]# -1 inside colon
print(y)
```
